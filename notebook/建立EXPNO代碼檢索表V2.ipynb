```python
import pandas as pd

df = pd.read_excel('..\data\combined_excel_D.xlsx',index_col=0)
df
```

```python
database = df.drop_duplicates().dropna(axis=0)
database
```

```python
database = database[['受益人_ERP','產品名_ERP','開狀人_ERP','出口代碼','TXDAT','開狀銀行_電文','from']]
database
```

```python
cols = database.columns
new_cols = []
for i in cols:
    if '_' in i:
        new_cols.append(i.split('_')[0])
    elif i == '出口代碼':
        new_cols.append('EXPNO')
    else:
        new_cols.append(i)
new_cols
```

```python
database.columns = new_cols
database
```

```python
database.to_csv('../data/對應表/EXPNO對應表V2.csv')
```
