```python
import pandas as pd
import re

def preprocess(x):
    x = str(x) #轉str
    x = re.sub('[\u4e00-\u9fa5]', '', x) # 去除中文
    x = re.sub(r'[^\w\s]','',x) # 去除標點符號
    x = x.replace('\n', '').replace('\r', '').replace('\t', '') # 去除換行符號
    return str.strip(x)
```

```python
df = pd.read_excel('../data/combined_excel.xlsx',index_col=0)
df['59'] = df['59'].apply(lambda x:preprocess(x).upper())
df = df[['59','EXPNO']].dropna(axis=0)
df.head()
```

```python
def g(x):
    word1list = ['CORP','CO','CORPORATION']
    word2 = 'DIVISION'
    for word1 in word1list:
        if x.find(word1) == -1:
            continue
        start = x.find(word1)+len(word1)
        end = x.find(word2)
    return x[start:end]
df.loc[df['59'].str.contains('DIVISION')]
df
```

```python
a = df.loc[df['59'].str.contains('DIVISION')]
a['59'] = a['59'].apply(lambda x:g(x))
a['EXPNO'] = a['EXPNO'].apply(lambda x:x.replace(' ','').strip())
a = a.drop_duplicates()
a
```

```python
a.to_csv('../data/寶典/DIVSION映射表.csv')
```
