```python
idx_list = []
for idx in df.index:
    if len(df.loc[idx,'根據產品預測代號']) == 6: # 如果只有一個產品
        if str(df.loc[idx,'推薦公司事業部'][0]) == str(df.loc[idx,'利用公司名稱預測公司代號'][0]): #檢查第一碼
            idx_list.append(idx)
```

```python
df = df.loc[idx_list]
df = df[df['預測產品使用方式']=='rule']
df = df.drop_duplicates(subset=['預測產品(取長度最長)'])
df
```
