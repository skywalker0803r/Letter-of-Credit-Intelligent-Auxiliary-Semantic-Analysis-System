```python
idx_list = []
for idx in df.index:
    # 條件1 只有一個代碼 且 第一碼跟公司第一碼一模一樣
    cond1 = (len(df.loc[idx,'根據產品預測代號']) == 6)&(str(df.loc[idx,'推薦公司事業部'][0]) == str(df.loc[idx,'利用公司名稱預測公司代號'][0]))
    # 條件2 預測出 []
    cond2 = (len(df.loc[idx,'根據產品預測代號']) == 2)
    if cond1 | cond2:
        idx_list.append(idx)
print(len(idx_list))
```

```python
df = df.loc[idx_list]
print(df.shape)
df
```
